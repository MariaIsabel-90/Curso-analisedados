# Análise da renda per capta

```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
base_dados = pd.read_excel('Dados_Pib.xlsx')
base_dados.head()
```

```python
#olhar o pib médio de cada estado
base_dados.groupby(by = ['Territorialidades']).mean()
```

```python
#Sistema de grades

cor_fundo = '#191970 '
#o grafico
grid_graficos = sns.FacetGrid(base_dados, col = 'Territorialidades', hue = 'Territorialidades', col_wrap= 4);

#adicionar um gráfico de linha em cada local disponível
grid_graficos = grid_graficos.map(plt.plot, 'Ano', 'PIB per capita')

#adicionar uma sombra
grid_graficos = grid_graficos.map(plt.fill_between, 'Ano', 'PIB per capita', alpha = 0.2 ).set_titles('{col_name} Territoriedades')

grid_graficos = grid_graficos.set_titles('{col_name}')

#adicionar subtitulo
grid_graficos = grid_graficos.fig.suptitle(
    'Evolução da renda per capita por estado do Brasil \n Este relatório foi elaborado em um treinamento de python para análise de dados. Está disponível no canal de Youtube @Data Viking'
)

#ajustando
plt.subplots_adjust(top = 0.92)
```
