# Conjunto de dados de ações de empresas do ramo energético na B3

O objetivo é comparar as ações de empresas do ramo energético com as ações da Petrobras

```python
# Libs Necessárias

# Libs para Modelagem e Matrizes
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
base_dados = pd.read_excel('Dados_Empresas_Energia.xlsx')
```

```python
#valores de fechamento
base_dados.head()
```

```python
#series temporais, devemos colocar a data no index, já q vai ser o eixo x dos graficos
base_dados.set_index('Data',inplace=True);
```

```python
#Grafico

#estilo do grafico
plt.style.use('seaborn-darkgrid')

#tamanho da figura
plt.figure(figsize= (16,7))

#titulo
plt.title('Analise ações de empresas de Energia', loc= 'left', fontsize = 12, fontweight=600)

plt.xlabel('Período das ações')
plt.ylabel('Preço fechamento (R$)');

plt.text(base_dados.index[-1],base_dados['Petrobras'].tail(1), 'Petrobras' ,color='#008c4a', size = 'large')
#grafico de linhas
plt.plot(base_dados.index, base_dados['Petrobras'], color='#008c4a',linewidth=4,alpha=0.7)

#plotar todas as colunas
for coluna in base_dados.columns[1:]:
  plt.plot(base_dados.index, base_dados[coluna], color='gray',linewidth=2,alpha=0.7)
  plt.text(base_dados.index[-1],base_dados[coluna].tail(1), coluna ,color='gray', size = 'large')


```
